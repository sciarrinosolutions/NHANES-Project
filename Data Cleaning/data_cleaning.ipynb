#### Diet
The following features represent dietary intake over 2 days. The mean of each will be calculated to get an average daily dietary intake value.

* If both days were usual = mean of day 1 and day 2
* If only day 1 was usual = just day 1's value
* If only day 2 was usual = just day 2's value
* If neither day was usual = NaN

```python
day1_usual = df_copy["DR1_300"] == 2
day2_usual = df_copy["DR2_300"] == 2

nutrients = {
    "kcal":     ("DR1TKCAL", "DR2TKCAL"),
    "protein":  ("DR1TPROT", "DR2TPROT"),
    "fat":      ("DR1TTFAT", "DR2TTFAT"),
    "carbs":    ("DR1TCARB", "DR2TCARB"),
    "chol_diet":("DR1TCHOL", "DR2TCHOL"),
    "sug_diet": ("DR1TSUGR", "DR2TSUGR"),
    "caff":     ("DR1TCAFF", "DR2TCAFF"),
    "h2o":      ("DR1_320Z", "DR2_320Z"),
    "fiber":    ("DR1TFIBE", "DR2TFIBE"),
    "sat_fat":  ("DR1TSFAT", "DR2TSFAT"),
    "mono_fat": ("DR1TMFAT", "DR2TMFAT"),
    "poly_fat": ("DR1TPFAT", "DR2TPFAT"),
    "sodium_diet":   ("DR1TSODI", "DR2TSODI"),

}

for new_col, (d1_col, d2_col) in nutrients.items():
    d1 = df_copy[d1_col].where(day1_usual, other=np.nan)
    d2 = df_copy[d2_col].where(day2_usual, other=np.nan)
    df_copy[new_col] = pd.concat([d1, d2], axis=1).mean(axis=1, skipna=True)
```

#### Physical activity

https://pmc.ncbi.nlm.nih.gov/articles/PMC7719906/

* mod_pa_min_per_wk  : Total minutes per week of moderate-intensity leisure-time physical activity (LTPA). Moderate activity is defined as causing moderate increases in breathing or heart rate.

* vig_pa_min_per_wk  : Total minutes per week of vigorous-intensity LTPA.Vigorous activity causes large increases in breathing or heart rate.

* total_pa_mets_per_wk: MET-weighted total physical activity per week, combining moderate (3.5 METs) and vigorous (7.0 METs) activity. Accounts for the greater physiological demand of vigorous activity relative to moderate activity.

* meets_pa_guidelines : Binary indicator of whether the respondent meets WHO 2020 guidelines: >=150 min/wk moderate OR >=75 min/wk vigorous intensity LTPA (1=meets guidelines, 0=does not).

```python
# check counts as both minutes and hours are used
print(df_copy['PAD800'].value_counts().sort_index().head(30))
print(df_copy['PAD820'].value_counts().sort_index().head(30))
```

```python
print(df_copy['PAD790U'].unique())
print(df_copy['PAD810U'].unique())
```

```python
# Convert moderate frequency to weekly rate
df_copy['PAD790Q'] = np.where(df_copy['PAD790U'] == "b'W'", df_copy['PAD790Q'],
                     np.where(df_copy['PAD790U'] == "b'D'", df_copy['PAD790Q'] * 7,
                     np.where(df_copy['PAD790U'] == "b'M'", df_copy['PAD790Q'] / 4.33,
                     np.where(df_copy['PAD790U'] == "b'Y'", df_copy['PAD790Q'] / 52,
                     df_copy['PAD790Q']))))

# Convert vigorous frequency to weekly rate
df_copy['PAD810Q'] = np.where(df_copy['PAD810U'] == "b'W'", df_copy['PAD810Q'],
                     np.where(df_copy['PAD810U'] == "b'D'", df_copy['PAD810Q'] * 7,
                     np.where(df_copy['PAD810U'] == "b'M'", df_copy['PAD810Q'] / 4.33,
                     np.where(df_copy['PAD810U'] == "b'Y'", df_copy['PAD810Q'] / 52,
                     df_copy['PAD810Q']))))

# Compute minutes per week
df_copy['mod_pa_min_per_wk'] = df_copy['PAD790Q'] * df_copy['PAD800']
df_copy['vig_pa_min_per_wk'] = df_copy['PAD810Q'] * df_copy['PAD820']
```

```python
df_copy[['mod_pa_min_per_wk', 'vig_pa_min_per_wk']].describe()
```

```python
(df_copy[['mod_pa_min_per_wk', 'vig_pa_min_per_wk']].isna()).sum()
```

```python
df_copy[['mod_pa_min_per_wk', 'vig_pa_min_per_wk']].describe()
```

```python
# MET-weighted total physical activity per week
df_copy['total_pa_mets_per_wk'] = (
    (df_copy['mod_pa_min_per_wk'] * 3.5) +
    (df_copy['vig_pa_min_per_wk'] * 7.0)
)
```

```python
df_copy['total_pa_mets_per_wk'].describe()
```

```python
# Binary: meets WHO 2020 physical activity guidelines
df_copy['meets_pa_guidelines'] = (
    (df_copy['mod_pa_min_per_wk'] >= 150) |
    (df_copy['vig_pa_min_per_wk'] >= 75)
)

df_copy['meets_pa_guidelines'] = df_copy['meets_pa_guidelines'].where(
    df_copy[['mod_pa_min_per_wk', 'vig_pa_min_per_wk']].notna().all(axis=1)
).astype('Int64')  # keeps NA as <NA>, not 0
```

#### Alcohol use

```python
# Convert ALQ121 to drinking days per week based on codebook response categories
alq121_map = {
    0:  0,        # Never in the last year
    1:  7,        # Every day (7 days/week)
    2:  6,        # Nearly every day (~6 days/week)
    3:  3.5,      # 3 to 4 times a week
    4:  2,        # 2 times a week
    5:  1,        # Once a week
    6:  0.577,    # 2 to 3 times a month (2.5/month / 4.33)
    7:  0.231,    # Once a month (1/month / 4.33)
    8:  0.212,    # 7 to 11 times in last year (9/52)
    9:  0.087,    # 3 to 6 times in last year (4.5/52)
    10: 0.029,    # 1 to 2 times in last year (1.5/52)
}

df_copy['ALQ121'] = df_copy['ALQ121'].replace(alq121_map)

# Calculate drinks per week
df_copy['drinks_per_week'] = np.where(
    df_copy['ALQ121'] == 0,
    0,
    df_copy['ALQ121'] * df_copy['ALQ130']
)
```

#### Balance

```python
# Average two trials per condition (uses trial 1 if trial 2 is missing)
df_copy['romberg_c1_time'] = df_copy[['BAXTC11', 'BAXTC12']].mean(axis=1, skipna=True)
df_copy['romberg_c2_time'] = df_copy[['BAXTC21', 'BAXTC22']].mean(axis=1, skipna=True)
df_copy['romberg_c3_time'] = df_copy[['BAXTC31', 'BAXTC32']].mean(axis=1, skipna=True)
df_copy['romberg_c4_time'] = df_copy[['BAXTC41', 'BAXTC42']].mean(axis=1, skipna=True)
df_copy['romberg_c5_time'] = df_copy[['BAXTC51', 'BAXTC52']].mean(axis=1, skipna=True)

time_cols = ['romberg_c1_time', 'romberg_c2_time', 'romberg_c3_time',
             'romberg_c4_time', 'romberg_c5_time']

# Count conditions attempted
df_copy['romberg_conditions_attempted'] = df_copy[time_cols].notna().sum(axis=1)

# Fill unattempted conditions with 0, then divide by 5
df_copy['romberg_avg_time'] = (
    df_copy[time_cols].fillna(0).sum(axis=1) / 5
)
```

```python
(df_copy['romberg_avg_time'].isna()).sum()
```

```python
# Set to NaN for people who didn't attempt any conditions
df_copy['romberg_avg_time'] = np.where(
    df_copy['romberg_conditions_attempted'] == 0,
    np.nan,
    df_copy['romberg_avg_time'])
```

# Feature Engineering

```python
# Mean arterial pressure (MAP)

# Mean Arterial Pressure (MAP) formula calculates average blood pressure using systolic (SBP) and diastolic (DBP) readings
# MAP = 1/3*(SBP) + 2/3*(DBP)

df_copy['map'] = 1/3 * df_copy['bp_sys'] + 2/3 * df_copy['bp_di']
```

```python
df_copy['map'].describe()
```

```python
# Daily macronutrient percentages of total caloric intake
# Protein: 1g = 4kcal, Carbs: 1g = 4kcal, Fat, 1g = 9kcal https://caloriedata.org/calculators/macro-calculator

df_copy['carbs_pct'] = (df_copy['carbs'] * 4) / df_copy['kcal'] * 100
df_copy['fat_pct'] = (df_copy['fat'] * 9) / df_copy['kcal'] * 100
df_copy['protein_pct'] = (df_copy['protein'] * 4) / df_copy['kcal'] * 100
```

```python
df_copy[['carbs_pct', 'fat_pct', 'protein_pct']].describe()
```

```python
df_copy['fiber_per_1000kcal'] = df_copy['fiber'] / df_copy['kcal'] * 1000
```

```python
# ═══════════════════════════════════════════════════════
# MULTICOLLINEARITY AUDIT — ALL 12 TARGETS
# ═══════════════════════════════════════════════════════

from scipy.stats import spearmanr
import pandas as pd
import numpy as np

# ── Define all 12 blood test targets ────────────────────
# Add all targets and their clinical thresholds here
TARGETS = {
    "triglycerides":     150,
    "hba1c":             6.5,
    "vitamin_d":         20,
    "alt":               40,
    "albumin":           3.5,
    "alp":               120,
    "ast":               40,
    "bun":               20,
    "hdl_cholesterol":   40,
    "ldl_cholesterol":   130,
    "cholesterol_total": 200,
    "hscrp":             3,
}

# ── Define competing feature groups ─────────────────────
feature_groups = {
    "fiber": {
        "features":  ["fiber", "fiber_per_1000kcal"],
        "rationale": "per_1000kcal normalizes for total intake"
    },
    "fat_total": {
        "features":  ["fat", "fat_pct"],
        "rationale": "pct normalizes for total intake"
    },
    "carbs": {
        "features":  ["carbs", "carbs_pct"],
        "rationale": "pct normalizes for total intake"
    },
    "protein": {
        "features":  ["protein", "protein_pct"],
        "rationale": "pct normalizes for total intake"
    },
    "blood_pressure": {
        "features":  ["bp_sys", "bp_di", "map"],
        "rationale": "map is linear combination of bp_sys and bp_di"
    },
    "physical_activity": {
        "features":  ["mod_pa_min_per_wk", "vig_pa_min_per_wk",
                      "total_pa_mets_per_wk", "meets_pa_guidelines",
                      "sedentary_time_per_day"],
        "rationale": "mets combines mod and vig; sedentary is independent"
    },
    "fat_subtypes": {
        "features":  ["sat_fat", "mono_fat", "poly_fat", "fat_pct"],
        "rationale": "subtypes may add independent signal beyond total pct"
    },
    "calories": {
        "features":  ["kcal", "fat_pct", "carbs_pct", "protein_pct"],
        "rationale": "pct features absorb kcal information"
    }
}

# ── Run Spearman correlation against all targets ─────────
results = []

for target_name, threshold in TARGETS.items():
    if target_name not in df_copy.columns:
        continue

    for group_name, group_info in feature_groups.items():
        for feat in group_info["features"]:
            if feat not in df_copy.columns:
                continue

            combined = df_copy[[feat, target_name]].dropna()
            missing  = df_copy[feat].isna().mean() * 100

            if len(combined) < 100:
                continue

            corr, p = spearmanr(combined[feat], combined[target_name])

            results.append({
                "target":    target_name,
                "group":     group_name,
                "feature":   feat,
                "abs_corr":  abs(corr),
                "corr":      corr,
                "p_value":   p,
                "missing":   missing,
                "n":         len(combined)
            })

results_df = pd.DataFrame(results)

# ── Summary table — best feature per group per target ───
print("=" * 70)
print("BEST FEATURE PER GROUP PER TARGET")
print("=" * 70)

# For each group find which feature wins most often across targets
group_winners = {}

for group_name in feature_groups.keys():
    group_df = results_df[results_df["group"] == group_name]
    if group_df.empty:
        continue

    print(f"\n{'─'*70}")
    print(f"GROUP: {group_name}")
    print(f"Rationale: {feature_groups[group_name]['rationale']}")
    print(f"{'─'*70}")
    print(f"{'Target':<25} {'Winner':<25} {'Corr':>8} {'Runner-up':<25} {'Corr':>8}")
    print(f"{'-'*95}")

    win_counts = {f: 0 for f in feature_groups[group_name]["features"]}

    for target_name in TARGETS.keys():
        target_group = group_df[group_df["target"] == target_name]
        if target_group.empty:
            continue

        # Sort by absolute correlation
        sorted_group = target_group.sort_values("abs_corr", ascending=False)
        if len(sorted_group) < 1:
            continue

        winner     = sorted_group.iloc[0]
        runner_up  = sorted_group.iloc[1] if len(sorted_group) > 1 else None

        win_counts[winner["feature"]] = \
            win_counts.get(winner["feature"], 0) + 1

        runner_str = (f"{runner_up['feature']:<25} "
                      f"{runner_up['corr']:>8.4f}"
                      if runner_up is not None else "")

        # Flag if winner and runner-up are very close
        close = ""
        if runner_up is not None:
            diff = winner["abs_corr"] - runner_up["abs_corr"]
            close = " ≈" if diff < 0.02 else ""

        print(f"  {target_name:<23} {winner['feature']:<25} "
              f"{winner['corr']:>8.4f} {runner_str}{close}")

    # Overall winner for this group
    overall_winner = max(win_counts, key=win_counts.get)
    print(f"\n  → Wins across targets: "
          + ", ".join([f"{f}: {c}" for f, c in
                       sorted(win_counts.items(),
                              key=lambda x: x[1],
                              reverse=True)]))
    group_winners[group_name] = overall_winner
    print(f"  → Recommended to KEEP: {overall_winner}")

# ── Final recommendation ─────────────────────────────────
print("\n" + "=" * 70)
print("FINAL RECOMMENDATIONS")
print("=" * 70)

keep    = []
drop    = []

for group_name, group_info in feature_groups.items():
    winner   = group_winners.get(group_name)
    features = group_info["features"]

    for feat in features:
        if feat not in df_copy.columns:
            continue
        missing = df_copy[feat].isna().mean() * 100

        # Keep winner unless missingness is too high
        if feat == winner:
            if missing < 50:
                keep.append(feat)
                print(f"  KEEP: {feat:<30} "
                      f"(winner in {group_name}, {missing:.1f}% missing)")
            else:
                # High missingness — check runner up
                print(f"  WARNING: {feat:<26} won in {group_name} "
                      f"but {missing:.1f}% missing — review manually")
        else:
            drop.append(feat)

# Remove duplicates — a feature might appear in multiple groups
keep = list(set(keep))
drop = [f for f in list(set(drop)) if f not in keep]

print(f"\nFeatures to KEEP ({len(keep)}): {sorted(keep)}")
print(f"Features to DROP ({len(drop)}): {sorted(drop)}")

# ── Add to DROP_COLS ─────────────────────────────────────
print("\n" + "=" * 70)
print("UPDATED DROP_COLS ADDITION")
print("=" * 70)
print("\nAdd these to your DROP_COLS list:")
print(f"REDUNDANT_FEATURES = {sorted(drop)}")
```
